```python
from map import MolMap
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metric = 'cosine'
method = 'tsne'
```

```python
mp = MolMap(split_channels=False,
            metric = metric,
            var_thr = 0.0001)
```

```python
mp.fit(method=method,verbose=2,random_state=1)
```

```python
mp.plot_scatter('./html')
```

```python
mp.plot_grid('./html')
```

```python
mp.save('./%s_%s.mp' % (metric, method))
```
